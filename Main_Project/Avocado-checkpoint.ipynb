```python
X = pd.get_dummies(df.drop(['Date','year','type'],axis=1),drop_first=True)
y = df['type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,stratify=y, random_state=42)
```

```python
scale = StandardScaler()
```

```python
X_train_scaled = scale.fit_transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_scaled,y_train)
```

```python
log_model.score(X_test_scaled,y_test)
```

```python
y_pred = log_model.predict(X_test_scaled)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(15,6),dpi=100)
pd.Series(index=X.columns,data = log_model.coef_[0]).sort_values(ascending=False).plot(kind='bar',color='blue')
plt.title('Coefficients graph');
```
