```python
df = df.replace(['Europe'] , 'Eastern Europe')
```

```python
df_mean = df.groupby(['Region'])[['Happiness Score']].mean()
df_mean.sort_values(by = 'Happiness Score')
```

```python
happinest_6 = df.loc[(df['Happiness Score'])>6]
happinest_6 = happinest_6.groupby(['Region'])[['Happiness Score']].count()
happinest_6.sort_values(by = 'Happiness Score')
```

```python
r_max_min = df.groupby(['Region'])[['Happiness Score']].max()-df.groupby(['Region'])[['Happiness Score']].min()
r_max_min.sort_values(by = 'Happiness Score')
```

```python
data_plot = df.set_index('Country')
data_plot = data_plot['Happiness Score']

fig, ax = plt.subplots(figsize = (20,40))                             
bar_plot = data_plot.plot.barh()                                       
bar_plot.set_title('Happiness Score of all countries')
bar_plot.set_xlabel('Happiness Score')
bar_plot.invert_yaxis()                                                
plt.xlim(0,8)                                                          

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20, 15)
sns.heatmap(df.corr(), cmap = 'Blues', annot = True)
plt.show()
```
